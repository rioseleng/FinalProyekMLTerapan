TF-IDF Vectorizer digunakan untuk menemukan representasi fitur yang penting dari setiap kategori judul kursus. Alat ini dari library scikit-learn akan mengubah nilai-nilai tersebut menjadi vektor dengan menggunakan metode fit_transform dan transform, serta melakukan pemecahan data menjadi bagian-bagian yang lebih kecil secara langsung.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(stop_words='english')

tf.fit(course_list['course_title'])
```

Transformasi data pada kolom course_title menjadi bentuk verktor matriks.

```python
tfidf_matrix = tf.fit_transform(course_list['course_title'])
tfidf_matrix.shape
```

Mengubah bentuk vectorizer yaitu vektor menjadi bentuk matriks.

```python
tfidf_matrix.todense()
```

```python
import numpy as np
```

Melakukan perhitungan derajat kesamaan atau similatiry degree antar judul kursus dengan teknik cosine similarity menggunakan library scikit-learn.

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_sim
```

Mengubah matriks cosine similarity menjadi bentuk dataframe antar judul kursus

```python
cosine_sim_df = pd.DataFrame(
    cosine_sim, index=course_list.course_title, columns=course_list.course_title)
print('Shape:', cosine_sim_df.shape)

cosine_sim_df.sample(10, axis=0)
```

Melakukan pendefinisian fungsi course_recommendations untuk menampilkan hasil rekomendasi kursus berdasarkan kesamaan course organization dan course difficulty dari sebuah kursus.

```python
def course_recommendations(course_list, similarity_data=cosine_sim_df, items=course_list[['course_title', 'course_organization', 'course_difficulty']], k=5):
    index = similarity_data.loc[:,course_list].to_numpy().argpartition(range(-1, -k, -1))
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    closest = closest.drop(course_list, errors='ignore')
    return pd.DataFrame(closest).merge(items).head(k)
```
